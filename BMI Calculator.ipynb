# Adult BMI Calcualtor

From CDC.gov:

"Body mass index (BMI) is a person’s weight in kilograms divided by the square of height in meters. BMI is an inexpensive and easy screening method for weight category—underweight, healthy weight, overweight, and obesity."

"BMI does not measure body fat directly, but BMI is moderately correlated with more direct measures of body fat. Furthermore, BMI appears to be as strongly correlated with various metabolic and disease outcome as are these more direct measures of body fatness."

Learn more here: https://www.cdc.gov/healthyweight/assessing/bmi/adult_bmi/index.html



Image source: https://sahyadrihospital.com/blog/difference-between-overweight-and-obese/

```python
age = int(input("Enter your age: "))
if age <20:
    print("Sorry, this calculator is best used for adults ages 20 and above.")
else:
    height = float(input("Enter your height (in inches): "))
    weight = float(input("Enter your weight (in pounds): "))
```

```python
BMI = 703 * (weight/(height**2))
BMI = float("{:.1f}".format(BMI))
```

```python
if BMI <18.5:
    bmi_classification = "Underweight"
elif BMI <= 24.9:
    bmi_classification = "Normal Weight"
elif BMI <= 29.9:
    bmi_classification = "Overweight"
elif BMI <= 34.9:
    bmi_classification = "Obese"
elif BMI <= 39.9:
    bmi_classification = "Severely Obese"
else:
    BMI >= 40
    bmi_classification = "Morbidly Obese"
```

```python
from prettytable import PrettyTable
x = PrettyTable()
```

```python
x.field_names = ["Age", "Height (inches)", "Weight (pounds)", 
                               "BMI", "BMI Classification"]
x.add_row([age, height, weight, BMI, bmi_classification])

print(x.get_string(title="Adult BMI Calculator Results"))
```

```python
if bmi_classification == "Underweight":
    print("Your BMI is",BMI, "indicating your weight is in the", bmi_classification, "category for adults of your height.")
    print("Discuss your BMI category with your healthcare provider as BMI may relate to your overall health and well-being.")
    print("Your healthcare provider might  determine possible reasons for being underweight and recommend support or treatment.")
elif bmi_classification == "Normal Weight":
    print("Your BMI is",BMI, "indicating your weight is in the", bmi_classification, "category for adults of your height.")
elif bmi_classification == "Overweight":
    print("Your BMI is",BMI, "indicating your weight is in the", bmi_classification, "category for adults of your height.")
    print("Discuss your BMI category with your healthcare provider as BMI may relate to your overall health and well-being.")
    print("Your healthcare provider might determine possible reasons for being overweight and recommend support or treatment.")
    print("Having excess weight can increase risk for chronic conditions, such as high blood pressure, type 2 diabetes, and high cholesterol.")
elif bmi_classification == "Obese":
    print("Your BMI is",BMI, "indicating your weight is in the", bmi_classification, "category for adults of your height.")
    print("Discuss your BMI category with your healthcare provider as BMI may relate to your overall health and well-being.")
    print("Your healthcare provider might determine possible reasons for obesity and recommend support or treatment.")
    print("Having obesity can increase risk for chronic conditions, such as high blood pressure, type 2 diabetes, and high cholesterol.")
elif bmi_classification == "Severely Obese":
    print("Your BMI is",BMI, "indicating your weight is in the", bmi_classification, "category for adults of your height.")
    print("Discuss your BMI category with your healthcare provider as BMI may relate to your overall health and well-being.")
    print("Your healthcare provider might determine possible reasons for severe obesity and recommend support or treatment.")
    print("Having severe obesity can increase risk for chronic conditions, such as high blood pressure, type 2 diabetes, and high cholesterol.")
else:
    bmi_classification == "Morbidly Obese"
    print("Your BMI is",BMI, "indicating your weight is in the", bmi_classification, "category for adults of your height.")
    print("Discuss your BMI category with your healthcare provider as BMI may relate to your overall health and well-being.")
    print("Your healthcare provider might determine possible reasons for morbid obesity and recommend support or treatment.")
    print("Having morbid obesity can increase risk for chronic conditions, such as high blood pressure, type 2 diabetes, and high cholesterol.")
```
